```python
import pandas as pd
from textblob import TextBlob
```

```python
#task1
def sentimentAnalyzer(text):
    x = TextBlob(text)
    score = x.sentiment.polarity

    if score < -0.2:
        return "Negative"
    elif score > 0.2:
        return "Positive"
    else:
        return "Neutral"
```

```python
#task2
print(sentimentAnalyzer('exciting '))
print(sentimentAnalyzer('happy'))
print(sentimentAnalyzer('good'))
print(sentimentAnalyzer('rich'))
print(sentimentAnalyzer('smile'))
print("-------------------")
print(sentimentAnalyzer('sad '))
print(sentimentAnalyzer('disappointed'))
print(sentimentAnalyzer('bad'))
print(sentimentAnalyzer('poor'))
print(sentimentAnalyzer('anger'))
print("-------------------")
print(sentimentAnalyzer('food '))
print(sentimentAnalyzer('animal '))
```

```python
#task3
path = "//Amazon_Unlocked_Mobile.csv"
data = pd.read_csv(path)
data
```

```python
Apple = data[data['Product Name'] == 'Apple iPhone 4s 8GB Unlocked Smartphone w/ 8MP Camera, White (Certified Refurbished)' ]
Apple
```

```python
print("Rows:",Apple.shape[0])
print("Columns:",Apple.shape[1])
```

```python
Apple["Review_Length"] = Apple["Reviews"].astype(str).apply(len)
Apple
```

```python
shortest = Apple["Review_Length"].min()
longest = Apple["Review_Length"].max()
average = Apple["Review_Length"].mean()

print("Shortest Review Length:", shortest)
print("Longest Review Length:", longest)
print("Average Review Length:", average)
```

```python
#task4
Apple["Sentiment"] = Apple["Reviews"].astype(str).apply(sentimentAnalyzer)
Apple
```

```python
#task5
sentiment_counts = Apple["Sentiment"].value_counts()
sentiment_counts
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
sentiment_counts.plot(kind="bar", color=["blue","black","gray"])
plt.title("Sentiment Distribution for Apple iPhone 4s 8GB Unlocked Smartphone")
plt.xlabel("Sentiment")
plt.ylabel("Number of Reviews")
plt.xticks(rotation=0)
plt.show()
```

```python
neg_reviews = Apple[Apple["Sentiment"]=="Negative"]["Reviews"].astype(str)
neg_words = neg_reviews.str.lower().str.split()
neg_words_exploded = neg_words.explode()
neg_words_filtered = neg_words_exploded[neg_words_exploded.str.len() > 3]
top_neg_words = neg_words_filtered.value_counts().head(20)
top_neg_words
```

```python
from nltk.corpus import stopwords
import nltk

nltk.download('stopwords')
neg_reviews = Apple[Apple["Sentiment"]=="Negative"]["Reviews"].astype(str)
neg_words = neg_reviews.str.lower().str.split()
neg_words = neg_words.explode()
stop_words = set(stopwords.words("english"))
neg_words_clean = neg_words[(neg_words.str.len() > 3) & (~neg_words.isin(stop_words))]
neg_top_words = neg_words_clean.value_counts().head(20)
neg_top_words
```
